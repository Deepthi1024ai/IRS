```python
# Method: Run-Length Encoding
def compress(text):
    compressed = ""
    count = 1

    # Loop through the text
    for i in range(1, len(text)):
        if text[i] == text[i - 1]:
            count += 1
        else:
            compressed += str(count) + text[i - 1]
            count = 1

    # Add the last character group
    compressed += str(count) + text[-1]
    return compressed


# Write sample input to file
with open("input.txt", "w") as f:
    f.write("AAGGTTT66")

# Read input and compress
with open("input.txt", "r") as f:
    data = f.read()
    compressed_data = compress(data)

# Print result
print("Original:", data)
print("Compressed:", compressed_data)

# Save compressed output to file
with open("output.txt", "w") as f:
    f.write(compressed_data)
```

```python
def locate(sub, main):
    size_main = len(main)
    size_sub = len(sub)
    print("Give String :", main)
    print("Search String:", sub)
    print("------------------------")
    found_any = 0
    index = 0
    while index <= size_main - size_sub:
        k = 0
        while k < size_sub:
            if main[index + k] != sub[k]:
                break
            k = k + 1
        if k == size_sub:
            print("Substring exists at position", index)
            found_any = 1
        index = index + 1
    if found_any == 0:
        print("Substring not found in given string")

full_text = "AABAACAADAABAABA"
search_text = "AABA"
locate(search_text, full_text)
```
